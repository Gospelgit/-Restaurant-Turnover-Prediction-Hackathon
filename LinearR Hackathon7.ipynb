```
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
```

```
df_train = pd.read_csv('/content/sample_data/Train_dataset_(1).csv')
```

```
columns_to_process = ['Instagram Popularity Quotient',  'Overall Restaurant Rating', 'Live Music Rating', 'Ambience', 'Resturant Tier', 'Facebook Popularity Quotient', 'Value Deals Rating', 'Comedy Gigs Rating', 'Live Sports Rating']

# Compute the mean of specified columns
column_means = df_train[columns_to_process].mean()

# Replace null values with the mean in specified columns
df_train[columns_to_process] = df_train[columns_to_process].fillna(column_means)
```

```
columns_to_process = []

# Compute the mean of specified columns
column_medians = df_train[columns_to_process].median()

# Replace null values with the median in specified columns
df_train[columns_to_process] = df_train[columns_to_process].fillna(column_medians)
```

```
columns_to_encode = ['Cuisine', 'City', 'Restaurant Theme']

# Perform frequency encoding for each column
for column in columns_to_encode:
    # Calculate the frequency of each category
    frequency_map = df_train[column].value_counts(normalize=True)

    # Map the frequencies to the original categories
    df_train[column + '_Frequency_Encoded'] = df_train[column].map(frequency_map)

# Drop the original categorical columns if needed
df_train.drop(columns_to_encode, axis=1, inplace=True)
```

```
df_train['Opening Day of Restaurant'] = pd.to_datetime(df_train['Opening Day of Restaurant'])
reference_date = pd.Timestamp('1970-01-01')
df_train['Date_as_int'] = (df_train['Opening Day of Restaurant'] - reference_date).dt.days
```

```
df_train.info()
```

```
df_train.drop(['Opening Day of Restaurant'], axis=1, inplace=True)
```

```
df_train = pd.get_dummies(df_train, columns=['Endorsed By', 'Restaurant Location', 'Restaurant Type'], dtype = int)
```

```
# Compute the correlation matrix
correlation_matrix = df_train.corr()

# Select the feature of interest
feature_of_interest = 'Annual Turnover'

# Extract the correlations for the feature of interest
correlations = correlation_matrix[feature_of_interest]

# Sort the correlations to find the most related features
sorted_correlations = correlations.abs().sort_values(ascending=False)

print(sorted_correlations)
```

```
sel_features = ['Hygiene Rating', 'Instagram Popularity Quotient',  'City_Frequency_Encoded', 'Overall Restaurant Rating', 'Live Music Rating', 'Restaurant Zomato Rating', 'Value for Money', 'Resturant Tier', 'Date_as_int', 'Staff Responsivness', 'Facebook Popularity Quotient']

sel = df_train[sel_features]
```

```
LR = LinearRegression()
```

```
model = LR.fit(sel, df_train['Annual Turnover'])
```

```
model.score(sel, df_train['Annual Turnover'])
```
